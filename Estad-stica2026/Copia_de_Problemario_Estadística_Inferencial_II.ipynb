# Unidad 1. Regresión lineal simple y correlación

## Medidas de dispersión
Suma de los cuadrados de las desviaciones de los valores de $X$ con respecto a su media:
$$ S_{xx} = \sum_{i=1}^{n} (X_i - \bar{X})^2 $$
Suma de los productos de las desviaciones de los valores de $X$ y $Y$ con respecto a sus medias:
$$S_{xy} = \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})$$
Suma de los cuadrados de las desviaciones de los valores de $Y$ con respecto a su media:
$$ S_{xx} = \sum_{i=1}^{n} (Y_i - \bar{Y})^2 $$


## Coeficiente de correlación y de determinación
Coeficiente de correlación de Pearson:
$$ r = \frac{S_{xy}}{\sqrt{S_{xx} \cdot S_{yy}}} $$
Coeficiente de determinación:
$$ r^2 $$

## Recta de regresión ajustada
La regresión lineal ajustada se representa mediante estadísticos:
$$ \hat{Y} = b_0 + b_1 X $$
donde $\hat{Y}$ representa el valor de $Y$ obtenido mediante la recta de regresión ajustada (no la verdadera $Y$). Los estadísticos $b_0$ y $b_1$ se calculan de la siguiente manera:

$$b_1 = \frac{S_{xy}}{S_{xx}}$$

$$ b_0 = \bar{Y} - b_1 \bar{X} $$

## Cálculo de residuales
Residuales:
$$e_i = Y_i - \hat{Y}_i$$

## Sumas de cuadrados SS (Sum of Squares)
Suma de los Cuadrados de los Errores:
$$ SSE = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2 = \sum_{i=1}^n e_i^2$$
Suma total de cuadrados:
$$SST = \sum_{i=1}^{n} (Y_i - \bar{Y})^2$$
Suma de cuadrados de regresión:
$$SSR = \sum_{i=1}^{n} (\hat{Y_i} - \bar{Y})^2$$
- SST: Mide la variabilidad total de los datos observados.
- SSR: Mide la variabilidad de los datos que el modelo de regresión explica.
- SSE: Mide la variabilidad no explicada por el modelo (es decir, los residuos).


## Intervalo de confianza
Estadístico de prueba _t_:
$$t = \frac{b_1}{SE(b_1)}$$
Error estándar de $b_1$:
$$SE(b_1) = \frac{\sqrt{SSE / (n-2)}}   {\sqrt{S_{xx}} }$$
o
$$SE(b_1) = \frac{\sqrt{\sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2  / (n-2)}}   {\sqrt{S_{xx}} }$$

Intervalo de confianza para $b_1$:
$$b_1 - t_{\alpha/2} \cdot SE(b_1) \lt \beta_1 \lt b_1 + t_{\alpha/2} \cdot SE(b_1)$$
donde $n$ representa la cantidad de pares de datos.

## Comprobación de supuestos
Comprobar suposiciones:
- Test de shapiro a los residuales $e_i$: Para comprobar si la distribución es normal sobre la recta.
- Grafico $X$ vs $Y$: Para observar si los datos soportan la suposición de linealidad.
- Gráfico de residuales: Para observar si los datos soportan la suposición de linealidad, complementario al coeficiente de correlación
- Test de Breusch-Pagan: Para detectar heteroscedasticidad en regresion lineal

Test de Shapiro:<br>
`from scipy.stats import shapiro`<br>
Después, se obtiene el valor-p:<br>
`_, valor_p_sh = shapiro(data)`<br>
- $H_0$: Los datos siguen una distribución normal
- $H_1$: Los datos no siguen una distribución normal

Test de Breusch-Pagan:<br>
`from statsmodels.stats.api import het_breuschpagan`<br>
Después, se obtiene el valor-p:<br>
`_, valor_p_bp, _, _ = het_breuschpagan(residuales, X)`
- $H_0$: Hay homoscedasticidad
- $H_1$: Hay heteroscedasticidad

## ANOVA en regresión lineal

| Fuente de variación | Suma de cuadrados (SS) | Grados de libertad (df) | Promedio de los cuadrados (MS) | Estadístico F |
|---------------------|------------------------|--------------------------|--------------------------------|---------------|
| Regresión  | $ SSR$ | $p$ | $MSR = \frac{SSR}{p}$ | $F = \frac{MSR}{MSE}$|
| Error | $SSE$| $n - p - 1 $| $MSE = \frac{SSE}{n - p - 1} $|     |
| Total | $SST $| $n - 1$|             |        |

donde $p$ es el número de parámetros para la recta de regresión ajustada (en la regresión simple $p=1$). Las hipótesis son:
$$H_0: \beta_1 = 0$$
$$H_1: \beta_1 \ne 0 $$

1.   Establesca una variable dependiente ($Y$) y una variable independiente ($X$).
2.   Realice un diagrama de dispersión para estos datos.

5. Calcule el coeficiente de determinación e interprete el resultado.

```python
import pandas as pd
df=pd.DataFrame({
    "promedio_examenes_cortos":[59,92,72,90,95,87,89,77,76,65,97,42,94,62,91],
    "promedio_final":[64,84,77,80,77,81,80,84,80,69,83,40,78,65,90]
})
df
```

```python
#definir variables
x=df["promedio_examenes_cortos"]
y=df["promedio_final"]
```

```python
import matplotlib.pyplot as plt
#configuración general
plt.figure(
    figsize=(6,4), #tamaño (ancho,alto)
    dpi=100 #resolución
)
#definir variables
x=df["promedio_examenes_cortos"]
y=df["promedio_final"]
#gráfico de disperción
plt.scatter(
    x,y,
    marker="*",
    color="pink",
    edgecolor="#f527E4", #color de borde
    alpha=0.9, #transparencia
    s=60, #tamaño de cada punto
    label="promedio_final" #etiqueta

)
#Titulo
plt.title(
    "Gráfica de dispersión",
    fontsize=12, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiqueta eje x
plt.xlabel(
    "promedio_examenes_cortos",
    fontsize=10
)
#Etiqueta eje y
plt.ylabel(
    "promedio_final",
    fontsize=10
)
#Tamaño de ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)
#malla
plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #Ancho de linea
    alpha=0.5, #transparencia
    color="red"
)
#leyenda
plt.legend(
    fontsize=10,
    loc="best",
    frameon=False,
    #bbox_to_anchor=(0.5,-0.15)
)
#Guardar Imagen
plt.savefig(
    "dispersión",
    bbox_inches="tight"
)
```

###¿Los datos soportan la suposición de linealidad?
Si se cumple la suposición de linealidad

```python
4#Calcule el coeficiente de correlación e interprete el resultado.
from scipy.stats import pearsonr
r, valor_p=pearsonr(x,y) # p es prueba de hipotesis
# H0:P=0 no hay correlación
# H1: P\0 si hay correlación, correlacion:el cambio en una variable esta vinculado con el cambio de otra variable
print(f"coeficiente de correlación (r): {r:0.4f}")
print(f"valor_p: {valor_p:0.4f}")
```

Si existe una correlación positiva, entre promedios de exámenes cortos y la calificación final.

```python
#5 Calcule el coeficiente de determinación e interprete el resultado.
print(f"coeficiente de determinación:{r**2:0.4f}")
```

El coeficiente de determinación es 0.7475 lo que indica que el 74.75% de la variabilidad de la calificación final es explicada por el promedio de exámenes cortos

```python
#6 Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
y_calculada=modelo.predict(x_constante)
modelo.params
```

```python
#Grafico de dispersión
import matplotlib.pyplot as plt
#configuración general
plt.figure(
    figsize=(6,4), #tamaño (ancho,alto)
    dpi=100 #resolución
)

plt.scatter(
    x,y,
    marker="*",
    color="pink",
    edgecolor="#f527E4", #color de borde
    alpha=0.9, #transparencia
    s=60, #tamaño de cada punto
    label="promedio_final" #etiqueta

)
plt.plot(
    x,y_calculada,
    marker="o",
    color="blue",
    linewidth=2.2,
    linestyle="-",
    markersize=5,
    markerfacecolor="white",
    markeredgecolor="black",
    label="promedio_final"
)
#Titulo
plt.title(
    "Gráfica de dispersión",
    fontsize=12, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiqueta eje x
plt.xlabel(
    "promedio_examenes_cortos",
    fontsize=10
)
#Etiqueta eje y
plt.ylabel(
    "promedio_final",
    fontsize=10
)
#Tamaño de ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)
#malla
plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #Ancho de linea
    alpha=0.5, #transparencia
    color="red"
)
#leyenda
plt.legend(
    fontsize=10,
    loc="best",
    frameon=False,
    #bbox_to_anchor=(0.5,-0.15)
)
#Guardar Imagen
plt.savefig(
    "dispersión",
    bbox_inches="tight"
)
```

```python
#7 Intervalo de confianza
modelo.conf_int(alpha=0.05)
```

```python
#8 Residuales
residuales=modelo.resid
#Grafico de dispersión
plt.scatter(
    x, residuales,
    marker="*",
    color="pink",
    edgecolor="#f527E4", #color de borde
    alpha=0.9, #transparencia
    s=60, #tamaño de cada punto
    label="promedio_final" #etiqueta

)
plt.axhline(y=0, color="gray", linestyle="--") #linea horizontal
#Titulo
plt.title(
    "Gráfica de dispersión",
    fontsize=12, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiqueta eje x
plt.xlabel(
    "promedio_examenes_cortos",
    fontsize=10
)
#Etiqueta eje y
plt.ylabel(
    "promedio_final",
    fontsize=10
)
#Tamaño de ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)
#malla
plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #Ancho de linea
    alpha=0.5, #transparencia
    color="red"
)
#leyenda
plt.legend(
    fontsize=10,
    loc="best",
    frameon=False,
    #bbox_to_anchor=(0.5,-0.15)
)
#Guardar Imagen
plt.savefig(
    "dispersión",
    bbox_inches="tight"
)
```

Los residuales se distribuyen alrededor de 0 aunque se observa mayor dispersión en los valores altos del promedio de exámenes cortos, también podría existir ligera variación en la varianza.

```python
#9 Test de Shapiro
from scipy.stats import shapiro
estadistico, valor_p=shapiro (residuales)
print(f"valor_p: {valor_p:0.4f}")
```

El valor p es mayor a 0.05 por lo tanto no se rechaza H0,l los residuales siguen una distribución normal y se cumple el supuesto de normalidad del modelo de regresión.

```python
#10 Test de Breusch-Pagan
from statsmodels.stats.api import het_breuschpagan
estadistico_1,valor_p_1,estadistico_2,valor_p_2=het_breuschpagan(residuales,x_constante)
print(f"valor-p:{valor_p_1:0.4f}")
#H0:No hay heterocedasticidad (varianza uniforme)
#h1:Hay heterocedasticidad (varianza no uniforme)
#Como el valor p es mayor al nivel de significancia, entonces la h0 es verdadera

```

El valor p es mayor a 0.05 entonces no se rechaza H0, por lo tanto no hay evidencias de heterocedasticidad y se cumple el supuesto de homocedasticidad en el modelo de regresión.

```python
import pandas as pd
df=pd.DataFrame({
    "numero_interrupciones_trabajador":[5,10,10,15,15,20,20,25],
    "calificacion_trabajador_prueba_hostilidad":[58,41,45,27,26,12,16,3]
})
df
```

```python
#definir variables
x=df["numero_interrupciones_trabajador"]
y=df["calificacion_trabajador_prueba_hostilidad"]
```

###¿Los datos soportan la suposición de linealidad?
Sí apoyan la suposición de linealidad

```python
4#Calcule el coeficiente de correlación e interprete el resultado.
from scipy.stats import pearsonr
r, valor_p=pearsonr(x,y) # p es prueba de hipotesis
# H0:P=0 no hay correlación
# H1: P\0 si hay correlación, correlacion:el cambio en una variable esta vinculado con el cambio de otra variable
print(f"coeficiente de correlación (r): {r:0.4f}")
print(f"valor_p: {valor_p:0.4f}")
```

Si existe una correlación negativa entre las variables

```python
#5 Calcule el coeficiente de determinación e interprete el resultado.
print(f"coeficiente de determinación:{r**2:0.4f}")
```

El coeficiente de determinción obtenido indica que el 98.58% de la variación de la variable dependiente es explicada por la variable independiente

```python
#6 Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
y_calculada=modelo.predict(x_constante)
modelo.params
```

```python
import matplotlib.pyplot as plt
#configuración general
plt.figure(
    figsize=(6,4), #tamaño (ancho,alto)
    dpi=100 #resolución
)

#gráfico de disperción
plt.scatter(
    x,y,
    marker="*",
    color="pink",
    edgecolor="#f527E4", #color de borde
    alpha=0.9, #transparencia
    s=60, #tamaño de cada punto
    label="calificacion_trabajador_prueba_hostilidad" #etiqueta

)
#Titulo
plt.title(
    "Gráfica de dispersión",
    fontsize=12, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiqueta eje x
plt.xlabel(
    "numero_interrupciones_trabajador",
    fontsize=10
)
#Etiqueta eje y
plt.ylabel(
    "calificacion_trabajador_prueba_hostilidad",
    fontsize=10
)
#Tamaño de ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)
#malla
plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #Ancho de linea
    alpha=0.5, #transparencia
    color="red"
)
#leyenda
plt.legend(
    fontsize=10,
    loc="best",
    frameon=False,
    #bbox_to_anchor=(0.5,-0.15)
)
#Guardar Imagen
plt.savefig(
    "dispersión",
    bbox_inches="tight"
)
```

```python
#7 Intervalo de confianza
modelo.conf_int(alpha=0.05)
```

```python
#8 Residuales
residuales=modelo.resid
#Grafico de dispersión
plt.scatter(
    x, residuales,
    marker="*",
    color="pink",
    edgecolor="#f527E4", #color de borde
    alpha=0.9, #transparencia
    s=60, #tamaño de cada punto
    label="calificacion_trabajador_prueba_hostilidad" #etiqueta

)
plt.axhline(y=0, color="gray", linestyle="--") #linea horizontal
#Titulo
plt.title(
    "Gráfica de dispersión",
    fontsize=12, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiqueta eje x
plt.xlabel(
    "numero_interrupciones_trabajador",
    fontsize=10
)
#Etiqueta eje y
plt.ylabel(
    "calificacion_trabajador_prueba_hostilidad",
    fontsize=10
)
#Tamaño de ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)
#malla
plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #Ancho de linea
    alpha=0.5, #transparencia
    color="red"
)
#leyenda
plt.legend(
    fontsize=10,
    loc="best",
    frameon=False,
    #bbox_to_anchor=(0.5,-0.15)
)
#Guardar Imagen
plt.savefig(
    "dispersión",
    bbox_inches="tight"
)
```

Los residuos se distribuyen aleatoriamente alrededor de 0 y no muestran un patrón definido, por lo que se cumplen razonablemente los supuestos del modelo lineal.

```python
#9 Test de Shapiro
from scipy.stats import shapiro
estadistico, valor_p=shapiro (residuales)
print(f"valor_p: {valor_p:0.4f}")
```

El valor p es ligeramente mayor a 0.05 entonces no se rechaza H0, sin embargo al estar muy cerca de 0.05 la normalidad se cumple de manera marginal

```python
#10 Test de Breusch-Pagan
from statsmodels.stats.api import het_breuschpagan
estadistico_1,valor_p_1,estadistico_2,valor_p_2=het_breuschpagan(residuales,x_constante)
print(f"valor-p:{valor_p_1:0.4f}")
#H0:No hay heterocedasticidad (varianza uniforme)
#h1:Hay heterocedasticidad (varianza no uniforme)
#Como el valor p es mayor al nivel de significancia, entonces la h0 es verdadera
```

El valor p es mayor a 0.05 por lo tanto H0 no se rechaza entonces se cumple el supuesto de homocedasticidad.
